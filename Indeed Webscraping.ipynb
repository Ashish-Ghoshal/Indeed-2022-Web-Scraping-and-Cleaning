```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
```

Starting with Single page

```python
# function to extract html document from given url
def getHTMLdocument(url):
      
    # request for HTML document of given url
    response = requests.get(url)
      
    # response will be provided in JSON format
    return response.text
```

```python
#url_to_scrape = "https://practice.geeksforgeeks.org/courses/"
url_to_scrape = "https://www.indeed.com/jobs?q=data%20scientist%20%2420%2C000&l=New%20York&start=10&vjk=b0b6e7ad9951c1ea"
#conducting a request of the stated URL above
  
# create document
html_document = getHTMLdocument(url_to_scrape)
  
# create soap object
soup = BeautifulSoup(html_document, 'html.parser')

  
```

Job Titles

```python
def extract_job_title_from_result(soup): 
  jobs = []
  
  for div in soup.find_all(name="div", attrs={"class":"css-1xpvg2o e37uo190"}):
    print(div)
    for ele in div.find_all(name="a", attrs={"class":"jcs-JobTitle css-jspxzf eu4oa1w0"}):
      for a in ele.find_all(name="span"):
        jobs.append(ele["title"])
  return(jobs)
print(extract_job_title_from_result(soup))
```

locations

```python
def extract_location_from_result(soup): 
  locations = []
  spans = soup.findAll('span', attrs={'class': 'location'})
  for span in spans:
    locations.append(span.text)
  return(locations)
extract_location_from_result(soup)
```
